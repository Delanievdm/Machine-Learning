```python
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
scaler = StandardScaler()
```

```python
scaled_data = scaler.fit_transform(X)
```

```python
print(scaled_data)
```

```python
scaled_data_train, scaled_data_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.30)
```

```python
scaled_data_train
```

```python
scaled_data_test
```

```python
logisticregress = LogisticRegression()
```

```python
model1 = logisticregress.fit(scaled_data_train, y_train)
```

```python
neuralnetwork = MLPClassifier()
```

```python
model2 = neuralnetwork.fit(scaled_data_train, y_train)
```

```python
model1.score(scaled_data_test, y_test)
```

```python
model2.score(scaled_data_test, y_test)
```
